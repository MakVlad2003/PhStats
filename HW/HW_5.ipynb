## Задача 1.

Пусть $X_1,...,X_n$ выборка из некоторого распределения $\mathrm{P}$, причем ${\sf D} X_1=\sigma^2<+\infty$, и $\sigma$ неизвестно.
Рассмотрим статистику $S^2=\frac1n\sum\limits_{i=1}^n(X_i-\overline{X})^2$.

1. Докажите, что статистика $S^2$ равна $\overline{X^2}-\overline{X}^2$.

$S^2=\frac1n\sum\limits_{i=1}^n(X_i-\overline{X})^2 = \frac{1}{n} \sum\limits_{i=1}^n (X_i^2 - 2 X_i \overline{X} + \overline{X}^2) = \frac{1}{n} \sum\limits_{i=1}^n X_i^2 - 2\overline{X}\frac{1}{n} \sum\limits_{i=1}^n X_i + \overline{X}^2 = $
$ = \frac{1}{n} \sum\limits_{i=1}^n X_i^2 - 2\overline{X}^2 + \overline{X}^2 = \frac{1}{n} \sum\limits_{i=1}^n X_i^2 - \overline{X}^2 = \overline{X^2}-\overline{X}^2$


2. Является ли она состоятельной оценкой $\sigma^2$? Для ответа на этот вопрос воспользуйтесь утверждениями с лекции.

**Определение:** Оценка $\widehat{\theta}_n(X_1, ..., X_n)$ называется состоятельной оценкой $\tau(\theta)$, если $\widehat{\theta}_n(X_1, ..., X_n) \xrightarrow[n \to \infty]{P_\theta} \tau(\theta), \forall \theta \in \Theta$

Из ЗБЧ следует, что $\overline{X}$ сходится по вероятности к $E(X)$, а $\overline{X^2}$ к $E(X^2)$. Тогда:

$\displaystyle{\lim_{n \to \infty}} S^2 = \displaystyle{\lim_{n \to \infty}} (\overline{X^2}-\overline{X}^2) \xrightarrow[n \to \infty]{P_\theta} \displaystyle{\lim_{n \to \infty}} (E(X^2)-(E(X))^2) = Var(X) = \sigma^2$ - состоятельная оценка.

3. Является ли статистика $S^2$ несмещенной оценкой $\sigma^2$? Для ответа на этот вопрос вспомните различные варианты определения дисперсии.

**Определение:** Оценка $\widehat{\theta}(X)$ называется несмещенной оценкой величины $\tau(\theta)$, если $E_\theta (\widehat{\theta}(X)) = \tau(\theta), \forall \theta \in \Theta$.

*   $E(\widehat{\theta}_n) = \theta$ - было доказано на лекции. Т.о. оценка - несмещенная.
*   По ЗБЧ выборочное среднее $\overline{X}$ сходится по вероятности к $\theta$ при $n \rightarrow \infty$, поэтому данная оценка явлется сильно состоятельной.
*   То что оценка является асимптотически нормальной, следует из ЦПТ, т.е. при достаточно больших n распределение выборочного среднего $\overline{X}$ будет приближаться к нормальному с праметрами $\theta$ и $\sigma^2 / n$.

## Задача 5


Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \sigma^2)$. Известно, что в качестве оценки параметра $\sigma^2$ можно использовать следующие оценки $S^2, \frac{n}{n-1}S^2$. Проверьте эти оценки на несмещенность повторив эксперимент с семинара.

Реализуйте две функции, каждая из которых на вход берет несколько выборок, а на выход выдает массив оценок  
(первая функция - $S^2$, вторая функция - $\frac{n}{n-1}S^2$)  для каждой выборки.  
*Примечание:* У матрицы, заданной в виде `numpy.array` есть метод `var` вычисляющий выборочную дисперсию.

```python
def estimate_s2(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    return np.var(X, axis = 1)

def estimate_s2_corrected(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    n_samples, sample_size = X.shape
    return np.var(X, axis = 1) * (sample_size / (sample_size - 1))
```

Проверьте правильность своей реализации

```python
X = np.array([[1, 3, 3, 6], [2, 5, 1, 89]])
estimate_s2(X)

assert(sum(estimate_s2(X) != np.array([3.1875, 1399.6875])) == 0)
assert(sum(estimate_s2_corrected(X) != np.array((4.25, 1866.25))) == 0)
```

Задайте список оценок и разные параметры для отрисовки графика

```python
estimators = [
    (estimate_s2, r'$S^2$', 'purple', 0.1),
    (estimate_s2_corrected, r'$\frac{n}{n-1}S^2$', 'green', 0.2)
]
```

Создайте 500 выборок размера `sample_size`.

```python
sample_count = 500
sample_size = 100
X = np.array([np.random.normal(0, 1, size = (sample_size,)) for _ in range(sample_count)])
X.shape
```

Визуализируйте полученные оценки на одном графике. Средние значения по 500 экспериментам для каждой оценки нужно обозначить звездочками.



```python
plt.figure(figsize=(15, 0.7 * len(estimators)))

for estimator, name, color, y in estimators:
    E = estimator(X)
    plt.scatter(E, np.zeros(sample_count) + y, alpha=0.1,
                s=100, color=color, label=name)
    plt.scatter(E.mean(), y, marker='*', s=300,
                color='w', edgecolors='black')

    plt.vlines(1, -1, 1, color='r')
    plt.title('Размер выборки = %d' % sample_size)
    plt.yticks([])
    plt.legend()
    plt.xlim((0.8, 1.2))
    plt.ylim((-0.1, 0.2 * len(estimators)))

plt.show()
```

Напишите функцию построения таких графиков в общем виде: на вход функция может получать любые распределения (`distributions`) и любые оценки (`estimators`).




```python
def est_plot(distribution, estimators, sample_size, sample_count):
    '''
    Построение графика разброса реализаций оценок и их средних значений.

    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    sample_size -- размер выборок
    sample_count -- количество генерируемых выборок
    '''
    X = distribution.rvs(size=(sample_count, sample_size))

    plt.figure(figsize=(15, 0.7 * len(estimators)))

    for estimator, name, color, y in estimators:
        E = estimator(X)
        plt.scatter(E, np.zeros(sample_count) + y, alpha=0.1, s=100, color=color, label=name)
        plt.scatter(E.mean(), y, marker='*', s=300, color='w', edgecolors='black')

        plt.vlines(1, -1, 1, color='r')
        plt.title('Размер выборки = %d' % sample_size)
        plt.yticks([])
        plt.legend()
        plt.xlim((0.8, 1.2))
        plt.ylim((-0.1, 0.2 * len(estimators)))

    plt.show()
```

Постройте три графика, аналогичных верхнему, для разных размеров выборок (10, 100, 500).


```python
sample_size_list = (10, 100, 500)  # размеры выборок
sample_count = 500  # количество экспериментов

for sample_size in sample_size_list:
    distribution = sps.norm
    est_plot(distribution, estimators, sample_size, sample_count)
```

Изучите поведение среднего оценок из первого пункта при росте размера $n$ выборки - постройте график зависимости среднего оценки по всем экспериментам от размера выборки. Для вычисления зависимости нужно один раз сгенерировать выборки из достаточно большого размера и посчитать оценки по префиксам, используя функции из `numpy`.




```python
def mean_plot(distribution, estimators, n_grid, sample_count):
    '''
    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    n_grid -- массив значений размера выборки
    sample_count -- количество генерируемых выборок
    '''
    plt.figure(figsize = (20, 5))
    X = distribution.rvs(size = (sample_count, 300))

    for size in n_grid:
        plt.scatter(int(size), estimators[0][0](X[:, :int(size)]).mean(), color = 'blue', s = 10)
        plt.scatter(int(size), estimators[1][0](X[:, :int(size)]).mean(), color='red', s = 10)

    plt.xlabel('Размер выборки')
    plt.ylabel('Среднее значение оценки')
    plt.legend ((r'$S^2$', r'$\frac{n}{n-1}S^2$'))
```

```python
n_grid = np.linspace(10, 400, 100)
mean_plot(sps.norm, estimators, n_grid, sample_count)
```

 Какие из оценок являются асимптотически несмещёнными (т.е. $\forall \theta \in \Theta\colon \mathsf{E}_\theta \widehat{\theta} \to \theta$ при $n\to +\infty$)?

**Ответ:** Оценку $\frac{n}{n-1}S^2$ можно считать несмещенной, т.к. ее среднее значение равно истинному параметру, который она оценивает. В то же время оценка $S^2$ является асимптотически несмещенной, так как при небольших размерах выборки может наблюдаться значительное смещение относительно истинного параметра, однако по мере увеличения размера выборки это смещение уменьшается.

## Задача 6.

Пусть $X_1, \ldots, X_n$ &mdash; выборка из распределения $\mathcal{N}(a, \sigma^2)$. Визуализируйте свойства состоятельности и асимптотической нормальности для оценки $S^2$ дисперсии $\sigma^2$.


Зададим параметры эксперимента

```python
a = 0  # истинное значение параметра
sigma = 1
sample_size = 300  # размер выборок
sample_count = 500  # количество выборок
n_range = (np.arange(sample_size) + 1)  # размеры подвыборок
```

Сгенерируем множество выборок из распределения $\mathcal{N}(a, \sigma^2)$ (количество выборок - `sample_count`, размер каждой выборки - `sample_size`)

```python
# генерируем множество выборок
samples = sps.norm.rvs(size = (sample_count, sample_size))
```

Вычислим оценки в зависимости от размера выборки, усредняем по последней оси

```python
estimation = np.cumsum((samples - np.mean(samples, axis = 1, keepdims = True))**2, axis = 1) / np.arange(1, samples.shape[1] + 1)
```

Визуализируем для каждой выборки все оценки в зависимости от размера выборки.

```python
def est_plot(estimation, sample_count, sample_size, left=None, right=None, xlim = sample_size):
    '''
    estimation -- массив оценок от размера выборки
    sample_count -- количество генерируемых выборок
    sample_size -- размер каждой выборки
    left, rigth -- границы доверительного интеравала, нужны будут далее
    xlim = область по x
    '''
    plt.figure(figsize=(15, 7))
    for i in range(sample_count):
        plt.plot(np.arange(sample_size) + 1, estimation[i], color='blue', alpha=0.05)
    if type(left) and type(right) is np.ndarray:

        plt.plot(np.arange(sample_size) + 1, left, color='black')
        plt.plot(np.arange(sample_size) + 1, right, color='black')
        labels = [r'$\hat{\theta}$', r'$\theta$',
                  'Границы доверительного интервала']
        handels = [Line2D([0], [0], color='blue', lw=2),
                   Line2D([0], [0], color='red', lw=2),
                   Line2D([0], [0], color='black', lw=2)]

    else:
        labels = [r'$\hat{\theta}$', r'$\theta$']
        handels = [Line2D([0], [0], color='blue', lw=2),
                   Line2D([0], [0], color='red', lw=2),]

    plt.hlines(sigma, 0 ,sample_size, color='red')

    plt.title('Поведение оценки для разных реализаций')
    plt.xlabel('Размер выборки')
    plt.ylim((0, 5))
    plt.legend(handels, labels)
    plt.xlim((0, xlim));
```

```python
est_plot(estimation, sample_count, sample_size)
```

Добавим доверительный интервал для нашей оценки.

```python
right = sigma + 2 * sigma / np.sqrt(n_range)
left = sigma - 2 * sigma / np.sqrt(n_range)

est_plot(estimation, sample_count, sample_size, left = left, right = right)
```

Приблизим этот график в области малых размеров выборки (до 50).

```python
est_plot(estimation, sample_count, sample_size, left = left, right = right, xlim = 50)
```
